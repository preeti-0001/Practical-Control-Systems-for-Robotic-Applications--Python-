```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl

# Open-loop transfer function G(s) = 20 / (s(s+5))
num = [20]
den = [1, 5, 0]   # s^2 + 5s + 0
G = ctrl.tf(num, den)

# Unity feedback closed-loop system
T = ctrl.feedback(G, 1)

# Step response
t, y = ctrl.step_response(T)

# Extract second-order parameters
wn, zeta, poles = ctrl.damp(T, doprint=False)

# Performance metrics
info = ctrl.step_info(T)

print("Natural Frequency (wn):", wn[0])
print("Damping Ratio (zeta):", zeta[0])
print("\nStep Response Parameters:")
for k, v in info.items():
    print(f"{k}: {v}")

# Plot response
plt.figure()
plt.plot(t, y)
plt.title("Joint Position Step Response")
plt.xlabel("Time (s)")
plt.ylabel("Position")
plt.grid(True)
plt.show()
```
