```python
import control as ctrl
import matplotlib.pyplot as plt

# Time delay
delay = 0.15

# First-order Pade approximation
num_delay, den_delay = ctrl.pade(delay, 1)
Delay = ctrl.tf(num_delay, den_delay)

# G(s) = K / (s(s+3))
num = [1]
den = [1, 3, 0]
G = ctrl.tf(num, den)

# Open-loop with delay
G_approx = Delay * G

# Root locus
plt.figure()
ctrl.root_locus(G_approx, grid=True)
plt.title("Root Locus with Time Delay (Pade Approximation)")
plt.show()
```
